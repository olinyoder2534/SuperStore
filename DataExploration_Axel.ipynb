## Correlation matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'superstore'
correlation_mat = superstore.corr()

# Set the size of the heatmap
plt.figure(figsize=(15, 15))

# Create the heatmap with annotations
sns.heatmap(correlation_mat, annot=True, cmap='coolwarm')

# Display the plot
plt.show()
```

### Data Exploration Response vs. predictor variables!

```python
income_threshold = 170000

#Cap the extreme values
superstore['income_capped'] = superstore['Income'].apply(lambda x: min(x, income_threshold))

#Now, you can use 'income_capped' for analysis
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Response', y='income_capped', data=superstore)
plt.ylabel('Income')
plt.title('Response vs.Income ')
plt.show()
```

With this fifth box plot, we can explore the relationship between the quality of shelving location
(ShelveLoc) and Sales. We observe a slight positive relationship between the quality of shelving
location and Sales as “Bad quality” brings fewer sales, then “Medium quality”, and lastly “Good
quality” brings in more sales.
